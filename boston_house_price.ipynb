```python
data.drop(columns = ["ID"],inplace= True)
```

```python
## independent and dependent features

input_data = data.iloc[:,:-1]
output_data = data["Price"]
```

```python
input_data.head()
```

```python
output_data.head()
```

```python
## train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(input_data,output_data,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
## standardizing the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

```python
## print the coefficients and intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## on which parameters the model has been trained
regression.get_params()
```

```python
print(type(regression))
```

```python
## prediction with test data
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
## plot scatter plot for prediction
plt.scatter(y_test, reg_pred)
```

```python
residuals = y_test-reg_pred
```

```python
residuals
```

```python
## plot this residuals
sns.displot(residuals,kind= "kde")
```

```python
## sctter plot wrt prediction and residuals
plt.scatter(reg_pred, residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

```python
## R square and adjusted R square
from sklearn.metrics import r2_score
score = r2_score(y_test,reg_pred)
print(score)
```
